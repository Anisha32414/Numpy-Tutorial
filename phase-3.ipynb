### Advance operation and data handling 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure :- [rest_id,2021,2022,2023,2024]
sales_data=np.array([
[101,15000,17000,19500,2300],
[102,20000,22000,25000,3000],
[103,19000,24000,25000,2900],
[104,12000,15000,17500,19000],
[105,16500,17900,21000,25000]])
print("sales data of 5 cafes are :- \n",sales_data)
print("\n")

# print details of first 3 cafes:-

print("details of first 3 cafes are:- \n",sales_data[:3])

#print all data off excluding the rest_ids :- 

print("\n data without rest_ids :- \n",sales_data[:,1:])
```

```python
#print the total revenue earned of each year 

print("total revenue in each year :- \n",np.sum(sales_data[:,1:],axis=0))

#print total revenue earned by each cafe

print("\n total revenue earned by each cafe :- \n",np.sum(sales_data[:,1:],axis=1))

#print min revenue earned by each cafe

print("\n min revenue by each cafe :-\n",np.min(sales_data[:,1:],axis=1))

#print maximum revenue earned in a year 

print("\n maximum revenue earned is :-\n",np.max(sales_data[:,1:],axis=0))

#print average revenue of each cafe

print("\n average revenue of each cafe is :- \n",np.average(sales_data[:,1:],axis=1))
```

```python
c_sum=np.cumsum(sales_data[:,1:],axis=1)
print(c_sum)

plt.figure(figsize=(5,4))
plt.xlabel("years")
plt.ylabel("revenue")
plt.plot(np.mean(c_sum,axis=0))
plt.title("graph of revenue over years")
plt.grid(True)
plt.show()
```

```python
vector1=np.array([2,5,1,3])
vector2=np.array([8,6,2,4])

# vector addition:-
print("vector addition is :- ",vector1+vector2)

#vector multiplication
print("\n vector multiplication :-",vector1 * vector2)

# dot product 
print("\n dot product of vector1 and vector2 :-",np.dot(vector1,vector2))# sum of all terms after vector multiplication

# angle ---> theta ???
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("\n angle between vector1 and vector2 is :- ",angle)
```
